```redshift
DROP TABLE dw.fact_delivery CASCADE;

CREATE TABLE IF NOT EXISTS dw.fact_delivery (
  match_id          BIGINT NOT NULL REFERENCES dw.dim_match(match_id),
  innings_number    INT    NOT NULL, -- from staging
  over_number       INT    NOT NULL,
  ball_in_over      INT    NOT NULL,

  striker_id        BIGINT REFERENCES dw.dim_player(player_id),
  non_striker_id    BIGINT REFERENCES dw.dim_player(player_id),
  bowler_id         BIGINT REFERENCES dw.dim_player(player_id),

  batting_team_id   BIGINT REFERENCES dw.dim_team(team_id),
  bowling_team_id   BIGINT REFERENCES dw.dim_team(team_id),

  runs_off_bat      SMALLINT,
  extras            SMALLINT,
  total_runs        SMALLINT,
  is_boundary_4     BOOLEAN,
  is_boundary_6     BOOLEAN,
  is_wicket         BOOLEAN,
  wicket_kind       VARCHAR(64),
  fielder_names     VARCHAR(512),

  stg_file_hashkey  VARCHAR(256),
  ingest_date       DATE,

  PRIMARY KEY (match_id, innings_number, over_number, ball_in_over)
)
DISTKEY(match_id)
SORTKEY(match_id, innings_number, over_number, ball_in_over);
```
